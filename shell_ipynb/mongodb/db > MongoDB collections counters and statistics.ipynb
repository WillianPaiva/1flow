```python
from oneflow.core.models.nonrel import *
from datetime import datetime
```

```python
articles = Article.objects.all().count()
reads = Read.objects.all().count()
feeds = Feed.objects.all().count()
never_fetched = Feed.objects.filter(last_fetch=None).count()
subscriptions = Subscription.objects.all().count()
users = User.objects.all().count()

print '## Counters on', datetime.now()
print '- Articles:', articles
print '- Reads:', reads, 'merged:', reads - articles
print '- Feeds:', feeds, 'never fetched:', never_fetched
print '- Subscriptions:', subscriptions, 'merged:', subscriptions - feeds
print '- Users:', users
```
